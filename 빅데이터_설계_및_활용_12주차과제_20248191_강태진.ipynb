```python
raw_IL_df = pd.read_csv("/content/drive/MyDrive/IL.csv")
raw_IL_df.head()

raw_IL_df.info()

IL_total_cases = raw_IL_df[['tot_cases', 'date_updated']]
IL_dateindex_df = IL_total_cases.set_index("date_updated")
IL_dateindex_df.head()
```

```python
CA_population = 39_529_000
TX_population = 31_290_831
WA_population = 8_035_700
VA_population = 8_700_000
IL_population = 12_710_158
USA_population = 338_289_856

CA_rate = round((CA_population / USA_population),3)
TX_rate = round((TX_population / USA_population),3)
WA_rate = round((WA_population / USA_population),3)
VA_rate = round((VA_population / USA_population),3)
IL_rate = round((IL_population / USA_population),3)

print(CA_rate)
print(TX_rate)
print(WA_rate)
print(VA_rate)
print(IL_rate)
```

```python
CA_dateindex_df_index = pd.to_datetime(CA_dateindex_df.index)
TX_dateindex_df_index = pd.to_datetime(TX_dateindex_df.index)
WA_dateindex_df_index = pd.to_datetime(WA_dateindex_df.index)
VA_dateindex_df_index = pd.to_datetime(VA_dateindex_df.index)
IL_dateindex_df_index = pd.to_datetime(IL_dateindex_df.index)
CA_dateindex_df_index.dtype

final_df = pd.concat([CA_dateindex_df * CA_rate, TX_dateindex_df * TX_rate, WA_dateindex_df* WA_rate, VA_dateindex_df* VA_rate, IL_dateindex_df* IL_rate], axis=1)
final_df.columns = ['CA', 'TX', 'WA', 'VA', 'IL']
final_df.head()

final_df.plot.line(rot=45)
# 가장 완화된 주는 WA이다.
```
