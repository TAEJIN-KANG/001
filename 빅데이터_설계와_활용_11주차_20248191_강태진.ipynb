```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
file = '/content/drive/MyDrive/owid-covid-data.csv'
df = pd.read_csv(file)
df.info()
```

```python
locations = df['location']
print(locations)
```

```python
locations.unique()
```

```python
kor_df = df[df['location'] == 'South Korea']
kor_df.head()

```

```python
usa_df = df[df['location'] == 'United States']
usa_df.head()
```

```python
France_df = df[df['location'] == 'France']
France_df.head()
```

```python
kor_date_index_df = kor_df.set_index('date')
kor_date_index_df.head()
```

```python
usa_date_index_df = usa_df.set_index('date')
usa_date_index_df.head()
```

```python
France_date_index_df = France_df.set_index('date')
France_date_index_df.head()
```

```python
kor_total_cases = kor_date_index_df['total_cases']
kor_total_cases
```

```python
usa_total_cases = usa_date_index_df['total_cases']
usa_total_cases
```

```python
France_total_cases = France_date_index_df['total_cases']
France_total_cases
```

```python
kor_date_index_df.index
```

```python
usa_date_index_df.index
```

```python
France_date_index_df.index
```

```python
target_df = pd.DataFrame({
    'KOR' : kor_total_cases,
    'USA' : usa_total_cases
}, index = kor_date_index_df.index)
target_df.head()

target_df.plot.line(rot=45)
```

```python
target_df = pd.DataFrame({
    'KOR' : kor_total_cases,
    'FRANCE' : France_total_cases
}, index = kor_date_index_df.index)
target_df.head()

target_df.plot.line(rot=45)
```

```python
target_df['2021-01-01' : '2021-12-31'].plot.line(rot=45)
target_df['2022-01-01' : '2022-12-31'].plot.line(rot=45)
```

```python
target_df['2023-01-01' : '2023-12-31'].plot.line(rot=45)
target_df['2024-01-01' : '2024-12-31'].plot.line(rot=45)
```

```python
kor_population=kor_date_index_df['population']['2020-01-05']
usa_population=usa_date_index_df['population']['2020-01-05']
print('미국인구(',usa_population,') / 대한민국 인구(',kor_population,')')
```

```python
kor_population=kor_date_index_df['population']['2020-01-05']
France_population=France_date_index_df['population']['2020-01-05']
print('프랑스 인구(',France_population,') / 대한민국 인구(',kor_population,')')
```

```python
rate = round((usa_population / kor_population),2)
rate
```

```python
rate = round((France_population / kor_population),2)
rate
```

```python
target_df = pd.DataFrame({
    'KOR' : kor_total_cases * rate,
    'USA' : usa_total_cases
}, index = kor_date_index_df.index)

target_df.iloc[50:56]
```

```python
target_df = pd.DataFrame({
    'KOR' : kor_total_cases * rate,
    'FRANCE' : France_total_cases
}, index = kor_date_index_df.index)

target_df.iloc[50:56]
```
